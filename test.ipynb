```python
from dominantcolors import get_image_dominant_colors
from IPython.display import display
import numpy as np
from PIL import Image as pil_image

def display_color(rgb):
    arr = np.full([30, 60, 3], rgb, dtype='uint8')
    display(pil_image.fromarray(arr, 'RGB'))
```

```python
display(pil_image.open('examples/image.jpg'))
dominant_colors = get_image_dominant_colors('examples/image.jpg', 3)
```

```python
for color in dominant_colors:
    display_color(color)
```
